```python
df2 = df.copy()
df2.head(2)
```

```python
# dropping null value
df2.dropna(inplace = True)
```

```python
# creating a new varible movies for just movie type show
movies = df2[df2['type'] == 'Movie']
```

```python
# Calculating average movie duration 

movies['duration_minutes'] = movies['duration'].str.replace(' min', '', regex=False)
movies['duration_minutes'] = pd.to_numeric(movies['duration_minutes'], errors='coerce')
movies = movies.dropna(subset=['duration_minutes'])

average_duration = movies['duration_minutes'].mean()
print(f"Average movie duration: {average_duration:.2f} minutes")
```

```python
movies['duration_minutes'] = movies['duration'].str.replace(' min', '').astype(float)
sns.boxplot(data= movies , x='duration_minutes')
plt.title('Distribution of Movie Durations')
```

```python
tv_shows = df2[df2['type'] == 'TV Show']

tv_shows['seasons'] = tv_shows['duration'].str.replace(' Season', '', regex=False)
tv_shows['seasons'] = tv_shows['seasons'].str.replace('s', '', regex=False)
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=tv_shows, x='seasons', order=tv_shows['seasons'].value_counts().index[:10])
plt.title('TV Shows by Number of Seasons')
plt.xlabel('Seasons')
plt.ylabel('Count')
plt.show()
```

```python
tv_shows['season_type'] = tv_shows['seasons'].apply(lambda x: '1 Season' if x == '1' else 'More than 1')
season_counts = tv_shows['season_type'].value_counts()

plt.figure(figsize=(6, 6))
season_counts.plot.pie(autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'], startangle=90)
plt.title('TV Shows: 1 Season vs More than 1')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
df2['genre_split'] = df['listed_in'].str.split(',')
df2_exploded = df2.explode('genre_split')
df2_exploded['genre_split'] = df2_exploded['genre_split'].str.strip()
genre_counts = df2_exploded['genre_split'].value_counts()
print(genre_counts.head(10))
```

```python
genre_counts.head(10).plot(kind='barh', color='coral')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_directors = df2['director'].value_counts().head(10) 

# Plot
plt.figure(figsize=(10, 5))
top_directors.plot(kind='barh', color='skyblue')
plt.xlabel('Number of Titles')
plt.title('Top 10 Directors on Netflix')
plt.gca().invert_yaxis()  # To show highest at the top
plt.tight_layout()
plt.show()
```
